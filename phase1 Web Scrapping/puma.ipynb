```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://us.puma.com/us/en/search?q=shoe'
```

```python
HEADERS = ({'USER-AGENT':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'})
```

```python
Webpage = requests.get(url, headers=HEADERS)
```

```python
Webpage
```

```python
Webpage.content
```

```python
soup = BeautifulSoup(Webpage.content,'html.parser')
soup
```

```python
puma_titles=[]
for i in soup.find_all('h3',class_="w-full mobile:text-sm mobile:pr-0 font-bold text-base pr-5 line-clamp-2"):
    puma_titles.append(i.text)
    
    
puma_titles
```

```python
colors=[]
for i in soup.find_all('span',class_="group-hover:hidden group-focus-within:hidden text-puma-black-300 uppercase text-xs whitespace-nowrap"):
    colors.append(i.text)
    
    
colors
```

```python
Price=[]
for i in soup.find_all('span',class_="whitespace-nowrap text-base text-puma-red font-bold"):
    Price.append(i.text)
    
    
Price
```

```python
rating=[]
for i in soup.find_all('div', class_="flex items-center"):
    rating.append(i.text)
    
    
rating
```

```python
rating=[]
for i in soup.find_all('span',class_="sr-only"):
    rating.append(i.text)
    
    
rating
```

```python
rating=[]
for i in soup.find_all('div', class_="flex items-center"):
    rating.append(i.text.split('/')[0])
    
    
rating
```

```python
print(len(puma_titles),len(colors),len(Price),len(rating))
```

```python
max_length = max(len(puma_titles), len(colors), len(Price))

# Pad shorter arrays with None (or any placeholder value)
colors += [None] * (max_length - len(colors))
Price += [None] * (max_length - len(Price))
```

```python
print(len(puma_titles), len(colors), len(Price))
```

```python
import pandas as pd

df = pd.DataFrame({'Name': puma_titles , 'colors': colors, 'Price': Price})
df
```
