```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup

# Assuming you have a WebDriver installed and set up (e.g., ChromeDriver)
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get("https://www.skechers.com")

# Wait for the page to load or for elements to become available
# You may need to adjust the waiting time based on the page's loading speed
driver.implicitly_wait(10)

# Get the page source after it has been modified by JavaScript
page_source = driver.page_source

# Use Beautiful Soup to parse the page source
soup = BeautifulSoup(page_source, 'html.parser')

# Now, continue with your code to extract titles
titles = []
for i in soup.find_all('a', class_="link c-product-tile__title"):
    titles.append(i.text)



# Print the titles
print(titles)
```

```python
len(titles)
```

```python
colors = []
for i in soup.find_all('div', class_="c-product-tile__color-swatches__label"):
    colors.append(i.text)



# Print the titles
print(colors)
```

```python
len(colors)
```

```python
category = []
for i in soup.find_all('div',class_="c-product-tile__gender"):
    category.append(i.text)




print(category)
```

```python
price = []
for i in soup.find_all('div',class_="price"):
    price.append(i.text)




print(price)
```

```python
len(price)
```

```python
import pandas as pd

df = pd.DataFrame({'Name':titles , 'colors': colors, 'Price': price, 'category':category})
df
```
