```python
# Filling 0 for the woodland data since the data is not provided in the website

woodland_indices = result_merge2[result_merge2['Brand'] == 'Woodland'].index
result_merge2.loc[woodland_indices, 'Reviews'] = result_merge2.loc[woodland_indices, 'Reviews'].fillna(0)
result_merge2.loc[woodland_indices, 'Comfort'] = result_merge2.loc[woodland_indices, 'Comfort'].fillna(0)
result_merge2.loc[woodland_indices, 'Performance'] = result_merge2.loc[woodland_indices, 'Performance'].fillna(0)
result_merge2.loc[woodland_indices, 'Rating'] = result_merge2.loc[woodland_indices, 'Rating'].fillna(0)

result_merge2.head()
```

```python
result_merge2.isnull().sum()
```

```python
rating_null_indices = result_merge2[result_merge2['Rating'].isnull()].index

#Filling the rating column with the mean is not the right choice since its reviews are 0 so we have to fill those indices with 0

result_merge2.loc[rating_null_indices, 'Rating'] = 0
```

```python
result_merge2.isnull().sum()
```

Filling the null values in the performance column

```python
performance_null_indices = result_merge2[result_merge2['Performance'].isnull()].index

# Now we have to deal with the performance whose missing values we can fill by using rating column
result_merge2['Performance'] = result_merge2['Performance'].fillna(result_merge2.groupby('Rating')['Performance'].transform('mean'))
```

```python
result_merge2.isnull().sum()
```

Dealing with remaining null values in the performance column 

```python
performance_null_indices = result_merge2[result_merge2['Performance'].isnull()].index
performance_null_indices
```

```python
# Now still some nan values are left in the performance column since there isnt any performance values that corresponds to that star rating.
# So best way is to fill the star rating in the performance column 

result_merge2['Performance'].fillna(result_merge2['Rating'], inplace=True)
```

```python
result_merge2.isnull().sum()
```

Dealing with the null values in the comfort column

```python
# Now we have to deal with the comfort whose missing values we can fill by using rating column
result_merge2['Comfort'] = result_merge2['Comfort'].fillna(result_merge2.groupby('Rating')['Comfort'].transform('mean'))
```

```python
# Still some comfort nan values are left since corresponding values are not present so we applied same technique but this time groupby
# Performance column 
result_merge2['Comfort'] = result_merge2['Comfort'].fillna(result_merge2.groupby('Performance')['Comfort'].transform('mean'))
```

```python
result_merge2.isnull().sum()
```

```python
comfort_null_indices = result_merge2[result_merge2['Comfort'].isnull()].index
comfort_null_indices
```

```python
#Now only we are left with 6 nan values whose rating is 4.9 so best way is to fill the values corresponding to 4.8 rating data
value = result_merge2.loc[result_merge2['Rating'] == 4.8, 'Performance'].values
value
```

Using train test split 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y1, train_size=0.7, random_state = 0, shuffle=True)
```

```python
X_train.head()
```

Performing Scaling 

```python
# Create a MinMaxScaler instance
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler on the training data and transform both training and testing data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

Model 1 : XGBRegressor

```python
from xgboost import XGBRegressor

model1 = XGBRegressor(reg_alpha=0.1,subsample=0.9,learning_rate=0.05, n_estimators=500, colsample_bytree=0.9)
model1.fit(X_train_scaled, y_train)

# Prediction on training data
training_data_preds = model1.predict(X_train_scaled)
# R squared value --> more close to 1 means more accurate
from sklearn import metrics
train_r2_value = metrics.r2_score(y_train, training_data_preds)
print("For Training Data:")
print('R square value : ', train_r2_value)

from sklearn.metrics import mean_absolute_error
print('MAE : ', mean_absolute_error(y_train, training_data_preds))

# Prediction on testing data
testing_data_preds = model1.predict(X_test_scaled)
# R squared value --> more close to 1 means more accurate
test_r2_value = metrics.r2_score(y_test, testing_data_preds)
print("For Testing Data:")
print('R square value : ', test_r2_value)

from sklearn.metrics import mean_absolute_error
print('MAE : ', mean_absolute_error(y_test, testing_data_preds))
```

Model 2: RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

# Create a RandomForestRegressor instance
model2 = RandomForestRegressor(n_estimators=500, random_state=30)

# Fit the model on the scaled training data
model2.fit(X_train_scaled, y_train)

# Prediction on training data
training_data_preds_rf = model2.predict(X_train_scaled)
train_r2_value_rf = metrics.r2_score(y_train, training_data_preds_rf)

print("For Training Data with RandomForestRegressor:")
print('R square value : ', train_r2_value_rf)
print('MAE : ', mean_absolute_error(y_train, training_data_preds_rf))

# Prediction on testing data
testing_data_preds_rf = model2.predict(X_test_scaled)
test_r2_value_rf = metrics.r2_score(y_test, testing_data_preds_rf)

print("For Testing Data with RandomForestRegressor:")
print('R square value : ', test_r2_value_rf)
print('MAE : ', mean_absolute_error(y_test, testing_data_preds_rf))
```

Model 3: RandomForestRegressor with hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_absolute_error, r2_score

# Create a RandomForestRegressor instance
rf_model = RandomForestRegressor(random_state=42)

# Define the hyperparameters and their possible values to search
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500],
    'max_depth': [None, 10, 20, 30, 40],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2']
}

# Create a scorer using mean absolute error (you can adjust this based on your preference)
mae_scorer = make_scorer(mean_absolute_error, greater_is_better=False)

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, scoring=mae_scorer, cv=5, n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters from the grid search
best_params = grid_search.best_params_

# Use the best model for predictions
best_model = grid_search.best_estimator_

# Make predictions on training data
training_data_preds_rf_tuned = best_model.predict(X_train_scaled)
train_r2_value_rf_tuned = r2_score(y_train, training_data_preds_rf_tuned)

print("For Training Data with Tuned RandomForestRegressor:")
print('Best Hyperparameters:', best_params)
print('R squared value:', train_r2_value_rf_tuned)
print('MAE:', mean_absolute_error(y_train, training_data_preds_rf_tuned))

# Make predictions on testing data
testing_data_preds_rf_tuned = best_model.predict(X_test_scaled)
test_r2_value_rf_tuned = r2_score(y_test, testing_data_preds_rf_tuned)

print("For Testing Data with Tuned RandomForestRegressor:")
print('R squared value:', test_r2_value_rf_tuned)
print('MAE:', mean_absolute_error(y_test, testing_data_preds_rf_tuned))
```
